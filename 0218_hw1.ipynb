畫一個函數圖形


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


x = np.linspace(-10, 10, 400)
theta = np.linspace(0, 2 * np.pi, 400)
y = x + (np.exp(theta)-(2 * x * np.sin(theta)))/(np.sin(theta)-np.cos(theta))

plt.plot(x, y)

plt.title("Plot of \ny = x + (np.exp(theta)-(2 * x * np.sin(theta)))/(np.sin(theta)-np.cos(theta)")
plt.xlabel("x")
plt.ylabel("y")

plt.grid(True)
plt.show()
```
